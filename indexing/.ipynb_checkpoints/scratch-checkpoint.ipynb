```python
df = pd.read_csv('./data/pelabelan_2_balanced.csv')
df
```

```python
df_true_reference = df[df['status']=='yes'][['reference']]
df_true_reference = df_true_reference.reset_index()
df_true_reference
```

```python
df_true_reference = df_true_reference.drop(['index'], axis=1)
```

```python
df_true_reference
```

```python
df_true_reference.drop_duplicates()
```

```python
df2 = df_true_reference.copy()
df2.head(2)
```

```python
for index, row in df2.iterrows():
    print(f'Index: {index}, reference: {row.get("reference", 0)}')
    break
```

```python
unique_words = []

for index, row in df2.iterrows():
    phrase = row.get("reference", 0)
    words = phrase.split()
    for word in words:
        if word not in unique_words:
            unique_words.append(word)
```

```python
len(unique_words)
```

```python
index_table
```

```python
index_table = pd.DataFrame(unique_words)
```

```python
index_table.shape
```

Task:
for each unique words, find its occurence in df2

```python
series = []
for i, row in index_table.iterrows():
    unique_word = row.get(0, 0)
    indexes = []
    for j, row2 in df2.iterrows():
        phrase = row2.get("reference", 0).split()
        if str(unique_word) in phrase:
            indexes.append(j)
    series.append(indexes)
```

```python
index_table['indexes'] = series
```
